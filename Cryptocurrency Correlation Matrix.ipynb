```python
# Import necessary Python packages
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt

# Importing Top 10 cryptocurrencies by market capitalisation
BTC = pd.read_csv("BTC-USD.csv")
ETH = pd.read_csv("ETH-USD.csv")
BNB = pd.read_csv("BNB-USD.csv")
XRP = pd.read_csv("XRP-USD.csv")
ADA = pd.read_csv("ADA-USD.csv")
DOGE = pd.read_csv("DOGE-USD.csv")
MATIC = pd.read_csv("MATIC-USD.csv")
SOL = pd.read_csv("SOL-USD.csv")
DOT = pd.read_csv("DOT-USD.csv")
LTC = pd.read_csv("LTC-USD.csv")


data = [BTC, ETH, BNB, XRP, ADA, DOGE, MATIC, SOL, DOT, LTC]

for i in data:
    # Replacing Index with First Column
    i.set_index("Date", inplace=True)
    
    #Dropping all columns except closing price
    i.drop(i.columns.difference(['Close']), axis=1, inplace=True)
    
    
#Renaming column label
new_column_names = ["BTC", "ETH", "BNB", "XRP", "ADA", "DOGE", "MATIC", "SOL", "DOT", "LTC"]

for i, x in enumerate(data):
    x.columns = [new_column_names[i]]

    
# Combining dataframe of all cryptocurrencies
df = pd.concat([BTC, ETH, 
                BNB, XRP, 
                ADA, DOGE, 
                MATIC, SOL, 
                DOT, LTC], axis=1)


# Computing correlation between 10 cryptocurrencies
corr_matrix = df.corr()


# Plotting of Heatmap
fig, ax = plt.subplots(figsize=(10,10))
sn.heatmap(corr_matrix, annot=True)
plt.tick_params(axis="x", labeltop=True, labelbottom=False, top=True, bottom=False)
plt.show()
```
